```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

# Request the web page
site = "https://www.transfermarkt.com/premier-league/transfers/wettbewerb/GB1/plus/?saison_id=2023&s_w=&leihe=1&intern=0/"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site, headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page, 'html.parser')

# Initialize the list to store all team data
teams_data = []

# Iterate through each team's name tag
for team_headline in soup.find_all('h2', class_='content-box-headline'):
    team_name_tag = team_headline.find('a', title=True)
    if not team_name_tag:
        continue
    team_name = team_name_tag.get('title')

    # Find the 'in' and 'out' tables based on the team name
    next_sibling = team_headline.find_next_sibling('div', class_='responsive-table')
    in_table = next_sibling.find('table') if next_sibling else None
    out_table = next_sibling.find_next_sibling('div', class_='responsive-table').find('table') if next_sibling else None

    # Extract player names from each table
    in_players = [row.find_all('td')[0].text.strip() for row in in_table.find_all('tr')[1:]] if in_table else []
    out_players = [row.find_all('td')[0].text.strip() for row in out_table.find_all('tr')[1:]] if out_table else []

    # Save all the information
    teams_data.append({
        'Team': team_name,
        'In': in_players,
        'Out': out_players
    })

# Output the results
for team in teams_data:
    print(f"{team['Team']} -> IN PLAYERS:")
    for player in team['In']:
        print(player)
    print("\n---\n")  # Add a gap between the 'in' and 'out' players
    print(f"{team['Team']} -> OUT PLAYERS:")
    for player in team['Out']:
        print(player)
    print("\n==========\n")  # Separate each team's data for clarity
```

```python
# This code assumes that `teams_data` is already filled with the information as previously provided.

# Initialize an empty list to store the transfer data
transfers = []

# Iterate over each team's data
for team in teams_data:
    # Get the team name
    from_team = team['Team']

    # Iterate over the 'In' players for the current team
    for player in team['In']:
        # Split the player name from the short name/initials
        full_name = player.split(player[-1])[-1].strip() if player[-1].isupper() else player
        # Append the transfer data to the list
        transfers.append({
            'Player Name': full_name,
            'From Team': 'Unknown',  # We don't have the information about the previous team here
            'To Team': from_team
        })

    # Iterate over the 'Out' players for the current team
    for player in team['Out']:
        # Split the player name from the short name/initials
        full_name = player.split(player[-1])[-1].strip() if player[-1].isupper() else player
        # Append the transfer data to the list
        transfers.append({
            'Player Name': full_name,
            'From Team': from_team,
            'To Team': 'Unknown'  # We don't have the information about the next team here
        })

# Now, `transfers` contains all the transfer data. Let's print it out.
for transfer in transfers:
    print(f"Player Name: {transfer['Player Name']}, From Team: {transfer['From Team']}, To Team: {transfer['To Team']}")
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

# Request the web page
site = "https://www.transfermarkt.com/premier-league/transfers/wettbewerb/GB1/plus/?saison_id=2023&s_w=&leihe=1&intern=0/"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site, headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page, 'html.parser')

# Initialize dictionaries to store all 'in' and 'out' players for each team
in_players = {}
out_players = {}
teams_data = []

# Iterate through each team's name tag to collect 'in' and 'out' players
for team_headline in soup.find_all('h2', class_='content-box-headline'):
    team_name_tag = team_headline.find('a', title=True)
    if not team_name_tag:
        continue
    team_name = team_name_tag.get('title')

    # Find 'in' players table
    in_table_div = team_headline.find_next_sibling('div', class_='responsive-table')
    in_table = in_table_div.find('table') if in_table_div else None
    if in_table:
        in_players[team_name] = [row.find_all('td')[0].text.strip() for row in in_table.find_all('tr')[1:]]

    # Find 'out' players table
    out_table_div = in_table_div.find_next_sibling('div', class_='responsive-table') if in_table_div else None
    out_table = out_table_div.find('table') if out_table_div else None
    if out_table:
        out_players[team_name] = [row.find_all('td')[0].text.strip() for row in out_table.find_all('tr')[1:]]

# Create a list to store transfer details
transfers_list = []

# Compare 'in' players of each team with 'out' players of all other teams
for to_team, players_in in in_players.items():
    for player in players_in:
        for from_team, players_out in out_players.items():
            if player in players_out and from_team != to_team:
                transfers_list.append({
                    'Player Name': player,
                    'From Team': from_team,
                    'To Team': to_team
                })

# Print out the transfer list
for transfer in transfers_list:
    print(f"Player Name: {transfer['Player Name']}, From Team: {transfer['From Team']}, To Team: {transfer['To Team']}")
```
